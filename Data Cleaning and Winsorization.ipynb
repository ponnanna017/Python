```python
# winsorisation

import pandas as pd 
import scipy.stats as scp
import seaborn as sns
```

```python
df= pd.read_csv("C:/Users/ponna/Downloads/framingham.csv")
df.head()
```

```python
sns.boxplot(df)
```

```python
# trimming using boxplot method (upper and lower and bound)- The normal technique
#trims outliers based on a percentile value- Winsorization (we ought to give a percentile limit)
```

```python
df["BMI"]= scp.mstats.winsorize(df["BMI"], limits= 0.05) #limit of 95%
```

```python
sns.boxplot(df["BMI"])
```

```python
df= pd.read_csv("C:/Users/ponna/Downloads/framingham.csv")
```

```python
df["BMI"]= scp.mstats.winsorize(df["BMI"], limits= 0.01) #limit of 99%
```

```python
sns.boxplot(df["BMI"])
```

```python
#apllying on all columns 

df= pd.read_csv("C:/Users/ponna/Downloads/framingham.csv")
for i in df.columns:
    df[i]= scp.mstats.winsorize(df[i], limits=0.05)
```

```python
df.head()
```

```python
sns.boxplot(df["diaBP"]) #just to check if winsorization has worked, we try boxplots on various columns- (no outliers)
```

```python
sns.boxplot(df["glucose"]) #we notice a lot of outliers 
```

```python
df["glucose"]= scp.mstats.winsorize(df["glucose"], limits= 0.10) 
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
df= IterativeImputer(max_iter=10, verbose=0)
df.fit([[1,2],[2,4],[4,8],[np.nan,16]])
```

```python
test=[[np.nan,6],[6,np.nan],[np.nan,12],[13,np.nan]]
```

```python
print(df.transform(test))
```

```python
print(np.round(df.transform(test)))
```

```python
df1= IterativeImputer(max_iter=10, verbose=0)
df1.fit([[1,2],[3,6],[4,8],[10,20],[np.nan,22],[7,np.nan]])

#Verbosity flag, controls the debug messages that are issued as functions are evaluated. The higher, the more verbose.
#Can be 0, 1, or 2.

#Maximum number of imputation rounds to perform before returning the imputations computed during the final round.
#A round is a single imputation of each feature with missing values. 
```

```python
test1= test1=[[np.nan,4],[6,np.nan],[np.nan,6],[4,np.nan],[33,np.nan]]
print(np.round(df1.transform(test1)))
```
